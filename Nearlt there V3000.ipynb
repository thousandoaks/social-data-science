## Let's remove NaN

```python
camerareadyDataFrame=theDataFrame[['ID_initiative','Unnamed: 6','CMB','CMB.1']].dropna(inplace=False)
```

```python
camerareadyDataFrame
```

```python
camerareadyDataFrame['Strenghs']=camerareadyDataFrame[['ID_initiative','CMB']]['CMB'].apply(lambda x:str(x).replace(' ','').split(','))
camerareadyDataFrame['Stakeholders']=camerareadyDataFrame[['ID_initiative','Unnamed: 6']]['Unnamed: 6'].apply(lambda x:str(x).replace(' ','').split(','))
camerareadyDataFrame['Risks']=camerareadyDataFrame[['ID_initiative','CMB.1']]['CMB.1'].apply(lambda x:str(x).replace(' ','').split(','))
```

```python
camerareadyDataFrame
```

```python
3
```

```python
StakeholdersLong=camerareadyDataFrame[['ID_initiative','Stakeholders']].explode('Stakeholders')
StakeholdersLong['Value']=1
StakeholdersLongPivoted=StakeholdersLong.pivot_table(index='ID_initiative',columns='Stakeholders',values='Value').fillna(0).add_prefix('stakeholders_')
```

```python
StakeholdersLongPivoted
```

```python
StrenghsLong=camerareadyDataFrame[['ID_initiative','Strenghs']].explode('Strenghs')
StrenghsLong['Value']=1
StrenghsLongPivoted=StrenghsLong.pivot_table(index='ID_initiative',columns='Strenghs',values='Value').fillna(0).add_prefix('strenghts_')
```

```python
StrenghsLongPivoted
```

```python
RisksLong=camerareadyDataFrame[['ID_initiative','Risks']].explode('Risks')
RisksLong['Value']=1
RisksLongPivoted=RisksLong.pivot_table(index='ID_initiative',columns='Risks',values='Value').fillna(0).add_prefix('risks_')
```

```python
RisksLongPivoted
```

```python
RisksLongPivoted.columns
```

## Let's bring everything onto the same matrix

```python
theSuperDuperMatrix=pd.concat([StakeholdersLongPivoted,RisksLongPivoted],axis=1)
```

```python
theSuperDuperMatrix.corr()
```

```python
3
```

```python
corr_mat = theSuperDuperMatrix.corr().stack().reset_index(name="correlation")
```

```python
# Draw each cell as a scatter point with varying size and color
g = sns.relplot(
    data=corr_mat,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8),
)

# Tweak the figure to finalize
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor(".7")
```
